```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates

# Define your tasks, start times, and durations (in days or other time units)
tasks = ['Task 1', 'Task 2', 'Task 3', 'Task 4']
start_times = ['2023-09-30', '2023-10-01', '2023-10-03', '2023-10-05']
durations = [2, 4, 1, 3]  # Duration of each task in days

# Create a DataFrame to hold task information
df = pd.DataFrame({'Task': tasks, 'Start': pd.to_datetime(start_times), 'Duration': durations})

# Create a figure and axis for plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Create a bar for each task
for i, task in enumerate(df.itertuples()):
    ax.barh(task.Task, task.Duration, left=task.Start, align='center')

# Format the x-axis to show dates properly
ax.xaxis.set_major_locator(mdates.DayLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)  # Rotate the x labels for better readability

# Add labels and title
ax.set_xlabel('Date')
ax.set_ylabel('Tasks')
ax.set_title('Task Timetable')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.cm as cm
import numpy as np

# Define machines, tasks, start times, and durations (in arbitrary time units)
machines = ['Machine 1', 'Machine 2', 'Machine 3', 'Machine 4']
tasks = ['Task A', 'Task B', 'Task C', 'Task D']
start_times = [0, 2, 5, 7]  # Start times in time units (e.g., hours)
durations = [4, 3, 2, 5]  # Duration of each task

# Number of machines
num_machines = len(machines)

# Generate colors based on an integer index
cmap = cm.get_cmap('tab20', num_machines)  # Use a colormap (e.g., 'tab20') with distinct colors
colors = [cmap(i) for i in range(num_machines)]  # Get a unique color for each machine

# Create a DataFrame to hold task information
df = pd.DataFrame({
    'Machine': machines,
    'Task': tasks,
    'Start': start_times,
    'Duration': durations
})

# Create a figure and axis for plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Create a bar for each task with a different color from the distribution
for i, task in enumerate(df.itertuples()):
    ax.barh(task.Machine, task.Duration, left=task.Start, align='center', color=colors[i])
    # Add text inside the rectangle (task name centered)
    ax.text(task.Start + task.Duration / 2, i, task.Task, va='center', ha='center', color='black', fontweight='bold')

# Set labels and title
ax.set_xlabel('Time (hours)')
ax.set_ylabel('Machines')
ax.set_title('Machine Task Timetable')

# Set x-axis to start at zero and extend based on the task durations
ax.set_xlim(0, max(df['Start'] + df['Duration']))

# Show the plot
plt.tight_layout()
plt.show()


```

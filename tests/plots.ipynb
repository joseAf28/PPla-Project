```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates

# Define your tasks, start times, and durations (in days or other time units)
tasks = ['Task 1', 'Task 2', 'Task 3', 'Task 4']
start_times = ['2023-09-30', '2023-10-01', '2023-10-03', '2023-10-05']
durations = [2, 4, 1, 3]  # Duration of each task in days

# Create a DataFrame to hold task information
df = pd.DataFrame({'Task': tasks, 'Start': pd.to_datetime(start_times), 'Duration': durations})

# Create a figure and axis for plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Create a bar for each task
for i, task in enumerate(df.itertuples()):
    ax.barh(task.Task, task.Duration, left=task.Start, align='center')
    ax.text(task.Start + pd.Timedelta(task.Duration / 2, 'D'), i, f'{task.Duration} days', ha='center', va='center')

# Format the x-axis to show dates properly
ax.xaxis.set_major_locator(mdates.DayLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)  # Rotate the x labels for better readability

# Add labels and title
ax.set_xlabel('Date')
ax.set_ylabel('Tasks')
ax.set_title('Task Timetable')

# Show the plot
plt.tight_layout()
plt.show()
```
